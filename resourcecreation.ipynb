```python
import ipywidgets as widgets
from IPython.display import clear_output

qualification_data = {}
start_time_data = {}

# Function to handle button click event for adding qualifications
def add_qualification(_):
    qualification = qualification_widget.value
    shift_type = shift_type_widget.value
    start_time = starttime_widget.value
    percentage = percentage_widget.value

    if qualification not in qualification_data:
        qualification_data[qualification] = []

    qualification_data[qualification].append({"shift_type": shift_type, "starttime": start_time, "percentage": percentage})




    clear_output(wait=True)
    display_qual_widgets()
    display_qualifications()

# Function to display qualification widgets
def display_qual_widgets():
    display(qualification_widget)
    display(shift_type_widget)
    display(starttime_widget)
    display(percentage_widget)
    display(add_qual_button)


# Function to display stored qualification data
def display_qualifications():
    total_percentage = sum(data['percentage'] for shift_list in qualification_data.values() for data in shift_list)

    print("Qualification Data:")
    for qualification, shift_list in qualification_data.items():
        print(f"Qualification: {qualification}")
        for data in shift_list:
            print(f"Shift Type: {data['shift_type']}")
            print(f"Percentage: {data['percentage']}%")
            print(f"starttime: {data['starttime']}")


    print(f"Total Percentage: {total_percentage}%")

# Create widgets for qualification input
qualification_widget = widgets.Dropdown(
    options=['C', 'C+E', 'B'],
    description='Qualification:'
)

shift_type_widget = widgets.Text(
    description='Shift Type:'
)
starttime_widget = widgets.Text(
    description='start time :'
)

percentage_widget = widgets.FloatText(
    description='Percentage:'
)



add_qual_button = widgets.Button(description='Add Qualification')
add_qual_button.on_click(add_qualification)



# Display qualification widgets and initial qualification data
display_qual_widgets()
display_qualifications()



```
