```python
import ipywidgets as widgets
from IPython.display import display, clear_output

qualification_data = {}

# Function to handle button click event for adding qualifications
def add_qualification(_):
    qualification = qualification_widget.value
    percentage = percentage_widget.value

    # Check if the total percentage exceeds 100
    total_percentage = sum(data['percentage'] for data in qualification_data.values())
    if total_percentage + percentage > 100:
        print("Total percentage exceeds 100. Please adjust percentages.")
        display_qual_widgets()
        display_qualifications()
        return

    qualification_data[qualification] = {"percentage": percentage}

    clear_output(wait=True)
    display_qual_widgets()
    display_qualifications()

# Function to display qualification widgets
def display_qual_widgets():
    display(qualification_widget)
    display(percentage_widget)
    display(add_qual_button)

# Function to display stored qualification data
def display_qualifications():
    print("Qualification Data:")
    for qualification, data in qualification_data.items():
        print(f"Qualification: {qualification}")
        print(f"Percentage: {data['percentage']}%")
        print()

# Create widgets for qualification input
qualification_widget = widgets.Dropdown(
    options=['C', 'C+E', 'B'],
    description='Qualification:'
)

percentage_widget = widgets.FloatText(
    description='Percentage:'
)

add_qual_button = widgets.Button(description='Add Qualification')
add_qual_button.on_click(add_qualification)

# Display qualification widgets and initial qualification data
display_qual_widgets()
display_qualifications()
```
